```python
sns.countplot(x="species", data=df)
plt.show()
```

```python
sns.countplot(x="sepal_length", data=df)
plt.show()
```

```python
sns.countplot(x="petal_width", data=df)
plt.show()
```

```python
sns.relplot(x="sepal_width", y="sepal_length", data=df)
plt.show()
```

```python
sns.lineplot(data=df, x="petal_width", y="petal_length")
plt.show()
```

```python
sns.histplot(data=df, x="sepal_length")
plt.show()
```

```python
sns.histplot(data=df, y="sepal_width")
plt.show()
```

```python
sns.countplot(x="petal_length", data=df)
plt.show()
```

```python
sns.countplot(x="petal_width", hue="species", data=df)
plt.show
```

```python
sns.boxplot(x="species", y="sepal_length", data=df)
plt.show()
```

```python
sns.violinplot(x="petal_length", y="sepal_width", data=df)
plt.show()
```

```python
sns.pairplot(df, hue="species")
plt.show()
```
