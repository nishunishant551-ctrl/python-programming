```python
sns.countplot(x="sex", data=tips)
plt.show()
```

```python
sns.countplot(x="smoker", data=tips)
plt.show()
```

```python
sns.countplot(x="day", data=tips)
plt.show()
```

```python
sns.relplot(x="tip", y="day", data=tips)
plt.show()
```

```python
sns.lineplot(data=tips, x="sex", y="total_bill")
plt.show()
```

```python
sns.histplot(data=tips, x="size")
plt.show()
```

```python
sns.histplot(data=tips, y="sex")
plt.show()
```

```python
sns.countplot(x="smoker", data=tips)
plt.show()
```

```python
sns.countplot(x="size", hue="sex", data=tips)
plt.show()
```

```python
sns.boxplot(x="total_bill", y="smoker", data=tips)
plt.show()
```

```python
sns.violinplot(x="total_bill", y="sex", data=tips)
plt.show()
```

```python
sns.pairplot(tips, hue="tip")
plt.show()
```

```python
sns.heatmap(tips.pivot_table(index="size", columns="time", values="total_bill", aggfunc="mean", observed=False), annot=True)
plt.show()
```
